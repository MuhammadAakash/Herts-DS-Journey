### Example 1: Using Boolean selection

We will create a random distribution of points, the select the points below a line.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate 500 uniformly distributed between 0 and 10 in x and y 
x = np.random.uniform(0, 10, 500)  
y = np.random.uniform(0, 10, 500)  
````

````python
fig, ax = plt.subplots()

# Do you recall what all of this is doing?
# If not, experiment with the parameters to 
# test how it works.
ax.scatter(x, y, color='#0000FF', marker='o', s=10) 
ax.tick_params(axis='both', labelsize=14)
ax.grid(True)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
````

### Now we want to draw a line through the points $(x_1,y_1)=(2,4)$ and $(x_2,y_2)=(4,8)$. 

The equation for a straight line is 
$$y = mx + b,$$
where $m$ is the slope and $b$ is the $y$-intercept.

To draw a line through two points, input the $x$ and $y$ coordinates for the points in  
$$(y - y_1) = \left(\frac{y_2 - y_1}{x_2 - x_1}\right)(x - x_1)$$
Doing some algebra gives<br> 
<div align="center">slope = $\left(\frac{y_2 - y_1}{x_2 - x_1}\right)$</div> <br>
and<br>
<div align="center">intercept = $y_1 - \left(\frac{y_2 - y_1}{x_2 - x_1}\right)x_1$.</div>

Inputting the points 1 and 2 above, the equation for the line is
$$y = 2x$$
(Here the slope is 2 and the $y$-intercept is 0.)

````python
fig, ax = plt.subplots()

# The same plotting code as before
ax.scatter(x, y, color='#0000FF', marker='o', s=10) 
ax.tick_params(axis='both', labelsize=14)
ax.grid(True)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# The coordinates of the two points as specified above
x_points = np.array([2,4]) # x coordinates of both points
y_points = np.array([4,8]) # y coordinates of both points

# Points along the line passing throught the points.
# The endpoints are beyond the boundaries of the plot.
x_line = np.array([-10,20])
y_line = 2*x_line

# Plot the points
ax.scatter(x_points, y_points, color='black', s=100)  

# Plot the line
ax.plot(x_line, y_line, color='black',linewidth=2)  
````

### Now we want to select all points below the line. 

We can do this with inequalities. The equation is 
$$y < 2x$$.

We can apply <b>boolean selection</b> to plot the desired points as red squares. 

````python
fig, ax = plt.subplots()

# The initial plotting code (as before)
ax.scatter(x, y, color='#0000FF', marker='o', s=10) 
ax.tick_params(axis='both', labelsize=14)
ax.grid(True)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# A mask of True and False values depending on whether the inequality 
# y < 2*x holds.
mask = y < 2 * x

# Plot red squares on the selected points
ax.scatter(x[mask], y[mask], color='#FF0000', marker='s', s=20) 

# Plot the points and line (as before)
ax.scatter(x_points, y_points, color='black', s=100)  
ax.plot(x_line, y_line, color='black',linewidth=2)  

# Make a data frame to help you inspect the results
df = pd.DataFrame({
    'x': x,
    'y': y,
    'mask': mask
})
print(df)
````

### Exercise 2: Stars from the European Space Agency's Gaia Mission

I have performed a query of ESA's Gaia database to pick 1000 randomly selected stars. Gaia provides photometry (measures of brightness) and astrometry (measures of the very slight motions of the stars on the sky).

The queried file is called: https://star.herts.ac.uk/~kuhn/DHV/gaia_random1000.parquet

Recall that a parquet file may be read with the following pandas function:
```python
gaia = pd.read_parquet('')
```

Read this file and inspect the contents.

````python
gaia = pd.read_parquet('gaia_random1000.parquet')
````

We want to calculate two quantities for each star: the star's distance (how far it is) and it's luminosity (how much light it produces).

Distance can be calculated with the astrometric parallax quantity, where<br>
<div align="center">distance = 1000/parallax.</div>
The unit for distance in astronomy is the 'parsec', where 1 parsec = $3.08\times10^{13}$ km.

The brightness of a star is included in its magnitude (phot_g_mean_mag). <b>Brighter stars have smaller magnitudes.</b> Distant objects appear fainter, so we need to correct for this using the followwing formula
<div align="center">absolute magnitude = phot_g_mean_mag + 5 log$_{10}$(parallax) - 10</div>



````python
# Calculates the quantities shown above
gaia['distance'] = 1000/gaia['parallax']
gaia['AbsoluteMagnitude'] = gaia['phot_g_mean_mag'] + 5*np.log10(gaia['parallax']) - 10

# It's helpful to convert some quantities to log. Each unit represents a factor of 10 increase.
gaia['log10_distance'] = np.log10(gaia['distance'])
````

````python
gaia
````

````python
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.scatter(gaia['log10_distance'], gaia['AbsoluteMagnitude'], alpha=1, color='black', s=10)

# Brighter stars have smaller absolute magnitudes. (See above)
plt.gca().invert_yaxis()

# Add titles and labels
plt.xlabel('Log10(distance)')
plt.ylabel('Absolute Magnitude')

# Show grid (optional)
plt.grid(True)
````
