### Plotting example 1: Drawing a curve

```python
# Here is an example plot showing a sinusoid.
# For the next exercise, you can use bits
# of this code for your own program.

import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots()

ax.grid(color='lightgray')

ax.plot(x, y, linewidth=2.0, color='black')

# Set plot range and tic parameters
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

ax.tick_params(labelsize=14) # increase label font size

# Labels the axes
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)

plt.show() # With jupyter notebooks there is no need to include this line 
```

### Exercise 1:

Now let $x(t)$ and $y(t)$ be a parametric equation. Plot the curve generated by $(x(t),y(t))$ for values of $t$ from 1 to $\pi$.

$x(t) = 2\sin(t)\cos(5t/6)$

$y(t) = 2\sin(t)\sin(5t/6)$

Be sure to optimise your plotting to make it look nice! (Your are *strongly* encouraged to borrow plotting elements from the example above.)

```python
t = np.linspace(0, 12*np.pi, 600)
x = 2*(np.sin(t))*np.cos(t*5/6)
y = 2*(np.sin(t))*np.sin(t*5/6)


fig, ax = plt.subplots()
ax.grid(color="black")
ax.plot(x, y, linewidth=2.0, color='black')
ax.set(xlim=(-2.5, 2.5), xticks=np.arange(-2, 3),ylim=(-2.5, 2.5), yticks=np.arange(-2, 3))
ax.tick_params(labelsize=12)

ax.set_xlabel("x", fontsize=12)
ax.set_ylabel("y", fontsize=12)

plt.show()
```

### Plotting example 2: Points on a sphere

```python
# A line of dots is drawn along the prime meridian (longitude = 0 degrees)

plt.figure()
plt.subplot(projection="aitoff")
plt.title("Aitoff")
plt.grid(True)

lats = np.array(np.linspace(-90,90,13)) * np.pi / 180
lons = np.array(np.zeros(13)) * np.pi / 180

plt.plot(lons, lats, 'o', markersize=6, color='blue')

plt.xlabel('Lon.')
plt.ylabel('Lat.')
```

### Exercise 2:

Draw an Aitoff projection of a globe with dots along the Tropic of Cancer (latitude = 23.4 degrees) and Tropic of Capricorn (latitude = -23.4 degrees).

```python
# Write your answer here.

# Draw an Aitoff projection of a globe with dots along the Tropic of Cancer (latitude = 23.4 degrees) and Tropic of Capricorn (latitude = -23.4 degrees).
plt.figure(figsize=(10,5))
plt.subplot(projection="aitoff")
plt.title("Aitoff")
plt.grid(True)

lats = np.array([23.4, -23.4]) * np.pi / 180
lons = np.array(np.linspace(-180,180,2)) * np.pi / 180

plt.plot(lons, lats, "o", markersize=4, color="red")
plt.xlabel("Lon.")
plt.ylabel("Lat.")
```
